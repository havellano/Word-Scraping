```python
#1) 
import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/Types_of_volcanic_eruptions'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
soup
print(soup.prettify())
```

```python
print(soup.get_text())
```

```python
#2)
Sur = soup.find('span', {'id': 'Surtseyan'}).find_next('p')
Sub = soup.find('span', {'id': 'Submarine'}).find_next('p')
Subg = soup.find('span', {'id': 'Subglacial'}).find_next('p')
```

```python
#3)
print(Sur.text)
print(Sub.text)
print(Subg.text)
```

```python
#4)
import re
modified_Sur = re.sub("eruption", "volcanic event", Sur.text, count=1)
print(modified_Sur)
```

```python
#5) 
with open("volcano_paragraph.txt", "w") as f:
    f.write(modified_Sur)
    f.write("\n\n")
    f.write(Sub.text)
    f.write("\n\n")
    f.write(Subg.text)
```
